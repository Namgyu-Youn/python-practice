```python
# Matrix Multiplication
# Matrix의 Sub-array가 column라고 생각하자.
# M[i][j]는 M[row][column]의 element로 접근한다.

def Multiply(A, B) :
# size : A(m*n), B(n*l)
  m, n, l = len(A), len(A[0]), len(B[0]) # 3,2,2

# C. Multiplied result is m*l Matrix
  Matrix = [[0]*l]*m

# Matrix-Multiplication : Repetition
# i,j,k = 1,1,1
  for i in range(m) :
    for j in range(l) :
      for k in range(n) :
        Matrix[i][j] += A[i][k]*B[k][j]

  return Matrix


Multiply([[1,4],[3,2],[4,1]], [[3,3],[3,3]]) # [[15,15],[15,15],[15,15]]
```

```python
# 올바른 괄호
# '(' : appendleft
# ')' : pop / return False
from collections import deque
def solution(S) :
  s = deque()
  for i in range(len(S)) :
    if S[i] == '(' : # Bracket is opened
      s.appendleft('(')
    elif S[i] == ')' : # Bracket may be closed
      if s == deque([]) : return False
      elif s[0] == '(' : s.pop()

  return True if s == deque([]) else False

print(solution(")()()"))
```

```python
# N개의 최소공배수
def solution(arr:list) :
  arr.sort()
  result = 1
  for i in range(len(arr)-1) :
    for j in range(i+1, len(arr)) :
      if arr[j] % arr[i] == 0 :
        arr[j] = (arr[j] // arr[i])

  for i in range(len(arr)) :
    result *= arr[i]

  return result

print(solution([2,6,8,14])) # 168
print(solution([20, 5, 6, 7, 3])) # 420
print(solution([2, 11, 4, 6, 8])) # 264
```

```python
# 소수 찾기
def prime_num(n) :
  if n < 2 : return False
  for i in range(2, int(n)) :
    if n % i == 0 : return False
  return True

from itertools import combinations
def solution(num:int) :
  nums = []
  for i in range(len(num)) :
    nums += list(map(''.join, combinations(num, i)))

  nums.remove('')
  for i in range(1, len(nums)) :
    nums[i] == int(nums[i])
  nums = list(map(int, nums))

  for char in nums :
    if prime_num(char) == False :
      nums.remove(char)

  return nums

print(solution("2120"))
```

```python
from itertools import combinations

def solution(num: str):
    nums = []
    for i in range(len(num) + 1):
        nums += list(map(''.join, combinations(num, i)))

    nums = list(filter(lambda x: x != "", nums))
    nums = list(map(int, nums))

    return nums

print(solution("011"))
print(solution("17"))
```
