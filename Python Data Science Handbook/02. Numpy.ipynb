# 배열 연산 : 브로드캐스팅


broadcasting은 axis가 달라도 element 사이의 연산 수행하게 해준다.

```python
# array의 기본적인 연산들은 element단위로 수행된다 (axis가 같아야만 가능)
x1 = np.array([0, 2, 4])
x2 = np.array([1, 3, 5])
print(x1 + x2)
print(x1 + 3) # broadcasting
```

```python
x1 = np.arange(4)
x2 = np.arange(4)[:, np.newaxis]
print(x1)
print(x2)
print(x1 + x2)
```

```python
# nums of row are same -> broadcasting
np.zeros((4, 4)) + np.array([1, 2, 3,4])
```

```python
a = np.arange(3) # [0, 1, 2]
b = np.arange(3)[:, np.newaxis] # [[0], [1]. [2]]
print(a + b)
```

Rules of Broadcasting
1. if dimensions of two araay are not same, lower-dimension array's left-most shape is filled by 1
2. If any dimension in shape is not same, array which shape is 1 is followed by other.
3. If any dimension is not same and not one, error

Ex1)

```python
M = np.ones((2, 3))
a = np.arange(3)
```

```python
# dimension of a is lower -> fill out 1
print(M.shape) # (2, 3)
print(a. shape) # (1, 3) by Rule1
```

```python
# 1st dimension is not same -> modify 1 to 2
print(M.shape) # (2, 3)
print(a. shape) # (2, 3)

# M = [[1, 1, 1], [1, 1, 1]]
# a = [[0, 1, 2], [0, 1, 2]]
print(M + a)
```

Ex2)

```python
a = np.arange(3).reshape((3, 1)) # (3, 1)
b = np.arange(3) # (3, )

print(a)
print(b)
```

```python
x1 = np.ones((2, 3)) # shape : (2,3)
x2 = np.arange(3) # shape: (,3)

# x2 : (,3) -> (1, 3) -> (2,3)
# result : [[1, 2, 3],[1, 2, 3]]

print(x1 + x2)
```

논리 연산자들의 Ufunc에서 이용 : 'np.biwise_' 이용

bit(이진수) 연산은 각 자리의 숫자에 대한 논리 작업을 수행한다.

```python
x1 = np.array([[2, 1, 0], [3, 1, 4]]) # [[010, 001, 000], [011, 001, 100]]
x2 = np.array([[1, 1, 2], [3, 0, 2]]) # [[001, 001, 010], [011, 000, 010]]

# and : &
print(x1 & x2) # same value : itself / not same : 0
# or : |
print(x1 | x2) # same value : itself / not same : 1
```

```python
# np.bitwise_xor : 1 if same, else 0
a = np.array([5, 3, 7], dtype=np.uint8)  # [00000101, 00000011, 00000111]
b = np.array([2, 6, 4], dtype=np.uint8)  # [00000010, 00000110, 00000100]

np.bitwise_xor(a, b)  # [00000111, 00000101, 00000011]
```

```python
# np.bitwise_not : if bit value 1->0 and 0->1
np.bitwise_not(x1)
```

```python
x = np.arange(7)
print(x > 3)
print(x < 5)

print((x > 3) & (x < 5))
```

```python
# numpy array 에 대한 논리 연산은 and, or 대신에 &. | 을 사용해야 한다.
x = np.arange(7)
print((x > 3) and (x < 5))
```

```python
# np.zeros_like() : np.zeros와 다르게, 변수값만큼 zero를 생성
np.random.seed(42)
x = np.random.randn(100)
bins = np.linspace(-5, 5, 20)
counts = np.zeros_like(bins) # bins의 각 구간(bin)에 속간 value의 갯수를 count함

i = np.searchsorted(bins, x)
np.add.at(counts, i, 1)
```

예제 : k - 최근접 이웃 알고리즘

```python
x = rand.rand(10, 2)
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()

# scatter : set each axis's interval
# scatter(x, y, .., size) : axis(x,y) and size
plt.scatter(x[:, 0], x[:, 1], s=100);
differences = x[:, np.newaxis, :] - x[np.newaxis, :, :]
```

```python
differences = x[:, np.newaxis, :] - x[np.newaxis, :, :]
sq_differences = differences ** 2 # 거리는 좌표값 차이의 제곱값
dist_sq = sq_differences.sum(-1)
print(dist_sq)
```

```python
nearest = np.argsort(dist_sq, axis=1) # np.argsort : NumPy에서 지원하는 sorting func
print(nearest)
```

```python
K = 2
nearest_partition = np.argpartition(dist_sq, K+1, axis=1)
plt.scatter(x[:,0], x[:,1], s=100)
for i in range(x.shape[0]) :
  for j in nearest_partition[i, :K+1] :
    plt.plot(*zip(x[j], x[i]), color='red')
```
